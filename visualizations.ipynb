```python
import os
import numpy as np
from IPython.display import HTML

from preprocessing import processed_shakespeare_data2
from HMM import unsupervised_HMM
from HMM_helper import (
    text_to_wordcloud,
    states_to_wordclouds,
    parse_observations,
    sample_sentence,
    visualize_sparsities,
    animate_emission
)
```

# Importing the dataset

```python
data, words_list, syllables, end_syllables, rhyme_dict, stress_dict = processed_shakespeare_data2()
```

```python
words_map = {}
for i in range(len(words_list)):
    words_map[words_list[i]] = i
```

# Visualization of the dataset

Visualize all of Shakepeare's sonnets as a wordcloud:

```python
text = open(os.path.join(os.getcwd(), 'data/shakespeare.txt')).read()
wordcloud = text_to_wordcloud(text, title='Shakespearean Sonnets')
```

# HMM with 10 hidden states

## Training an HMM

Now we train an HMM on our dataset. We use 10 hidden states and train over 100 iterations:

```python
hmm = unsupervised_HMM(data, 10, 100)
```

## Visualization of the sparsities of A and O

```python
visualize_sparsities(hmm, O_max_cols=500, O_vmax=0.01)
```

## Visualizing the wordcloud of each state

Below, we visualize each state as a wordcloud by sampling a large emission from the state:

```python
wordclouds = states_to_wordclouds(hmm, words_map)
```

## Visualization of the HMM

```python
anim = animate_emission(hmm, words_map, M=8)
HTML(anim.to_html5_video())
```
